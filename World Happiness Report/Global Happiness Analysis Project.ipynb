```python
df17 = pd.read_csv('2017.csv')
```

```python
df17.describe()
```

```python
df17.nunique()
```

```python
df17.drop('Whisker.high',axis =1,inplace = True)
df17.drop('Whisker.low',axis =1,inplace = True)
```

```python
display(df17.head(10)) # top 10 happiest countries
```

```python
display(df17.tail(10)) # top 10 least happy countries
```

```python
display(df17.corr()) #display the data
correlationMap = df17.corr() #calculate the correlation of the variables
plt.subplots(figsize=(10, 10)) #subplot 10*10 inch by inch image
sns.heatmap(correlationMap, annot=True, linewidths = 1.0) #plot the correlation as heat map
```

```python
plt.clf()  #to clear the current figure.

plt.figure(figsize=(10,20));  #plot the figure in 10*15 inch image

plt.subplot(4,1,1);      # 1st sbu-plot 
plt.scatter(df17['Happiness.Score'], df17['Economy..GDP.per.Capita.'], color='g');
plt.xlabel("Happiness.Score");
plt.ylabel("Economy/GDP per Capita");

plt.subplot(4,1,2);     # 2nd sbu-plot
plt.scatter(df17['Happiness.Score'], df17['Family'], color='b');
plt.xlabel("Happiness.Score");
plt.ylabel("Family");

plt.subplot(4,1,3);     # 3rd sub-plot
plt.scatter(df17['Happiness.Score'], df17['Health..Life.Expectancy.'], color='r');
plt.xlabel("Happiness.Score");
plt.ylabel("Health/Life Expectance");

plt.subplot(4,1,4);     # 4th sub-plot
plt.scatter(df17['Happiness.Score'], df17['Generosity'], color='y');
plt.xlabel("Happiness.Score");
plt.ylabel("Generosity");

plt.suptitle("Correlation between variables with happiness score",fontsize=15)
plt.tight_layout(pad = 4)
plt.show()
```

```python
df17.plot(kind = "scatter", x = "Happiness.Score", y = "Happiness.Rank" , 
            color = "blue", grid = True , figsize = (15,10), alpha = 0.5)

plt.xlabel("Happiness.Score")
plt.ylabel("Happiness.Rank")
plt.title("Happiness Score and Rank Proportional Relationship")
plt.show()
```
